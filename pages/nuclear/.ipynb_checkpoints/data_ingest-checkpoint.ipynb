```python
import requests, pandas as pd
from bs4 import BeautifulSoup

def fetch_wikipedia_reactors():
    url = "https://en.wikipedia.org/wiki/List_of_nuclear_reactors"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    # find the right table, parse into df...
    tables = pd.read_html(r.text)
    reactors = tables[0]  # adjust index as needed
    reactors.to_csv("data/reactors.csv", index=False)
    return reactors
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def fetch_nuclear_companies_from_wikipedia_category(output_csv: str,
                                                    category_url: str = "https://en.wikipedia.org/wiki/Category:Nuclear_power_companies") -> pd.DataFrame:
    """
    Scrapes the Wikipedia category page for 'Nuclear power companies' and returns
    a DataFrame with one column, 'company', listing each company name.
    
    Parameters
    ----------
    output_csv : str
        Path where the CSV of company names will be written.
    category_url : str
        URL of the Wikipedia category page to scrape.
    
    Returns
    -------
    pd.DataFrame
        DataFrame containing a single column 'company'.
    """
    resp = requests.get(category_url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    
    # The list of pages is under div#mw-pages → div.mw-category-group → ul
    company_elements = soup.select("#mw-pages .mw-category-group ul li a")
    companies = [el.text.strip() for el in company_elements]
    
    df = pd.DataFrame({"company": companies})
    df.to_csv(output_csv, index=False)
    return df

if __name__ == "__main__":
    df_companies = fetch_nuclear_companies_from_wikipedia_category(
        output_csv="data/nuclear_companies.csv"
    )
    print(f"Scraped {len(df_companies)} companies:")
    print(df_companies.head())
```
